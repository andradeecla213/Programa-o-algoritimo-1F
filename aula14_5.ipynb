```python
# Estrutura de decisões

# Escrever um programa que identifica 02 números e identifica a número maior

numero1 = int (input("Digite um número: "))
numero2 = int (input("Digite um número: "))

# Processamento e saída

if numero1 > numero2:
  print(f"O número {numero1} é maior que o número {numero2}")

elif numero1 == numero2:
  print(f"O número {numero1} é igual o número {numero2}")

else:
  print(f"O número {numero1} é menor que o número {numero2}")
```

```python
#Escrever um programa que o usúario digita o nome e a idade

#Identificar se é criança, adolecente e adulto

nome = str( input(" Digite seu nome: "))
idade = int( input(" Digite sua idade: "))

if idade <= 12:
  print (f"Oie {nome}, você é criança")
elif idade  <= 18:
  print (f"Oie {nome}, você é adolecente")
else:
  print(f"Oie {nome}, você é adulto")
```
