```python
# Tabuada

numero = int (input("Digite um número: "))

for i in range (1,11):
  print(f" {numero} x {i} = {numero*i}")
```

```python
# Fatorial

numero = int(input("Digite um número inteiro positivo: "))

fatorial = 1

for i in range(1, numero + 1): fatorial *= i

print(f"O fatorial de {numero} é {fatorial}")
```

```python
# Cedulas

cedulas = 100, 50, 20, 10, 5, 2, 1

try:
    valor_saque = int(input("Digite o valor do saque: "))

    if valor_saque <= 0:
        print("Valor inválido. Digite um valor maior que zero.")
    else:
        valor_restante = valor_saque
        print(f"Para sacar R$ {valor_saque}, o caixa entregará a menor quantidade de notas:")

        for nota in cedulas:
            quantidade = valor_restante // nota

            if quantidade > 0:
                print(f"{quantidade} notas de R$ {nota}")
                valor_restante %= nota

except:
  print("")
```
