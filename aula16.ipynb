```python
# projeto caixa eletronico

# você foi contratado para desesnvolver uma sistema simples de caixa eletronico
# O sistema deverá funcionar atraves de um menu interativo ultilizando "while"
#deve iniciar com um saldo inicial de 1000,00 e permitir que o usuario escolha diferentes operações até decidir sair do sistema

# ver saldo
# depositar
# sacar dinheiro
# sair

# regras:
# quando depositar somar ao saldo anterior;
# quando sacar subtrair do saldo anterior
# mostar o saldo para o usuario
```

```python
opcao = ""
saldo = ""
num_a = 0

print ("-----------------------MENU-----------------------")
print ("")
print ("Opção 1 - Ver saldo")
print ("Opção 2 - Depositar")
print ("Opção 3 - Sacar dinheiro")
print ("Opção 4 - sair")
print ("")
print ("==================================================")
print ("")

try:

  while opcao !=4:
    opcao = int(input("olá, escolha uma das opções para iniciar ou 4 para sair: "))
    if opcao == 1:
      print(f"Seu saldo atual é: R$1000")

    elif opcao == 2:
      num_a = float(input("Digite um número: "))
      print (f"Seu saldo é de {num_a + saldo}: ")

    elif opcao == 3:
      num_a = float(input("Digite um número: "))

      print (f"Seu saldo é de {saldo - num_a}: ")
except:
  print()
```
