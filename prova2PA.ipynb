```python
#par ou impar 1

numero = int( input("Digite um número: "))
resto = numero % 2

if resto == 0:
    print(f"Número {numero} é par")


else:
    print(f" O Número {numero} é impar")
```

```python
#Qual numero é maior 2

numero1 = int(input("Digite um número: "))
numero2 = int(input("Digite um número: "))

if numero1 > numero2:
  print (f"O número {numero1} é maior que o número {numero2}")

else:
  print (f"O número {numero2} é maior que o número {numero1}")
```

```python
#Nota do aluno 3


nota = int (input ("Digite sua nota: "))

if nota >= 7:
  print ("Você foi aprovado!! Parabéns")

elif nota >= 5 and nota < 7:
  print ("Você está de recuperação")

else:
  print ("Você está reprovado")
```

```python
#Definir se é criança, adolescente, adulto ou idoso 4

idade = int( input("Digite sua idade: ") )

if idade <= 12:
  print (f"Você é criança")

elif idade >= 13 and idade <= 17:
  print (f" Você é um adolescente")

elif idade > 18 and  idade < 59:
  print (f" Você é adulto")

else:
  print (f" Você é idoso")
```

```python
#negativo, positivo ou zero 6

numero = int(input("Digite um número: "))

if numero >0 :
  print (f"O número {numero} é positivo")

elif numero <0 :
  print(f"O número {numero} é negativo")

else:
  print("O número é zero")
```

```python
#calculadora 7

num1 = float(input("Digite o primeiro número: "))
operacao = input("Escolha a operação (+, -, *, /): ")
num2 = float(input("Digite o segundo número: "))

# Estrutura de decisão para realizar o cálculo
if operacao == "+":
    resultado = num1 + num2
    print (f"O resultado é {operacao}")
elif operacao == "-":
    resultado = num1 - num2
    print (f"O resultado é {operacao}")
elif operacao == "*":
    resultado = num1 * num2
    print (f"O resultado é {operacao}")
elif operacao == "/":
    if num2 != 0:
        resultado = num1 / num2
        print (f"O resultado é {operacao}")
    else:
        resultado = "Erro: Divisão por zero não permitida."
else:
    resultado = "Operação inválida!"

print(f"Resultado: {resultado}")
```
