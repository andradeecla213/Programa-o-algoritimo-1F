```python
# Operadores aritiméticos

# Pedir ao usúario para digitar 02 número

# Entrada

a = int (input("Digite um número: "))
b = int (input("Digite um número: "))

# Processamento

# Soma

soma = a + b
subtracao = a - b
multiplicacao = a * b
div_int = a // b
resto = a % b
potencia = a ** b
# Saída

print("")
print("-----------OPERADORES ARITIMÉTICOS-----------")
print("")
print("----------------------SOMA-------------------")
print("")
print(f"O resultado da soma do {a} mais {b} é igual a {soma}")
print("")
print("----------------------SUBTRAÇÃO-------------------")
print("")
print(f"O resultado da DIFERENÇA do {a} menos {b} é igual a {subtracao}")
print("")
print("----------------------MULTIPLICACAO-------------------")
print("")
print(f"O resultado da MULTIPLICACAO do {a} vezes {b} é igual a {multiplicacao}")
print("")
print("----------------------DIVISAO-------------------")
print("")
print(f"O resultado da DIVISAO INTEIRA do {a} dividido {b} é igual a {div_int}")
print("")
print("----------------------RESTO-------------------")
print("")
print(f"O resultado da DIVISAO do {a} dividido {b} o resto é igual a {resto}")
print("")
print("----------------------POTENCIA------------------")
print("")
print(f"O resultado da POTENCIA do {a} potenciacao de2 {b} é igual a {potencia}")
print("")
print("---------------------------------------------")
```
